```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ticket cleaning.csv')
```

```python
data
```

```python
grp=data.groupby('PaymentMode')['TicketPrice'].count().reset_index()
```

```python
grp
```

```python
plt.pie(grp['TicketPrice'],labels=grp['PaymentMode'])
```

```python
plt.bar(data['TicketType'],data['TicketPrice'])
```

```python
plt.scatter(data['PaymentMode'],data['TicketPrice'])
```

```python
plt.plot(data['BookingID'],data['TicketPrice'])
```

```python
fig,ax = plt.subplots(2,2,layout='constrained')
ax[0,0].bar(data['TicketType'],data['TicketPrice'])
ax[0,0].set_title('chart1')
ax[0,0].set_xlabel('Types')
ax[0,0].set_ylabel('Amount')
ax[0,0].tick_params('y',color='r',labelcolor='green',labelsize=10,right=True,labelright=True)
ax[1,1].plot(data['BookingID'],data['TicketPrice'])
ax[1,1].set_title('chart2')
ax[1,1].minorticks_on()
ax[1,1].tick_params('x',which='major',direction='inout',length=10,width=5)
ax[1,1].tick_params('x',which='minor',direction='in',length=6,width=3)
ax[0,1].scatter(data['PaymentMode'],data['TicketPrice'])
ax[0,1].set_title('chart3')
ax[1,0].pie(grp['TicketPrice'],labels=grp['PaymentMode'])
```

```python
fig,ax = plt.subplots(1,3,figsize=(18,8),edgecolor='r')
ax[0].bar(data['TicketType'],data['TicketPrice'])
ax[1].plot(data['BookingID'],data['TicketPrice'])
ax[2].scatter(data['PaymentMode'],data['TicketPrice'])
```

```python
fig,ax = plt.subplots(2,2,sharex=True)
ax[0,0].plot(data['BookingID'],data['TicketPrice'] ,label='1-5 month')
ax[0,0].set_xlim(1,5)
ax[0,0].legend()
ax[0,0].tick_params('x',color='r',labelcolor='green')
ax[1,1].plot(data['BookingID'],data['TicketPrice'],label='sales')
ax[0,1].plot(data['BookingID'],data['TicketPrice'])
ax[0,1].grid(True)
ax[0,1].spines['right'].set_visible(False)
ax[0,1].spines['top'].set_visible(False)
ax[0,1].spines['bottom'].set_visible(False)
ax[0,1].spines['left'].set_visible(False)
ax[1,1].set_ylim(500,1000)
ax[0,1].set_xlim(5,10)
ax[1,0].plot(data['BookingID'],data['TicketPrice'])
ax[1,0].set_xlim(10,15)
```
